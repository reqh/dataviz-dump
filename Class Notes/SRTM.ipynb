```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# The Space Shuttle Radar Topography Mission produced
# a unique dataset of accurate topography at 90 and 30m 
# resolution for most of the non-polar parts of Earth.
# https://dwtkns.com/srtm/ 

# http://srtm.csi.cgiar.org/wp-content/uploads/files/srtm_5x5/TIFF/srtm_19_04.zip
    
srtm = imageio.imread('srtm_19_04/srtm_19_04.tif')
```

```python
plt.hist(np.reshape(srtm, shape=(srtm.shape[0]* srtm.shape[1],)))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
```

```python
n2020 = pd.read_csv("Nat2020PublicUS.c20210506.r20210812.txt", header=None)
n2020
```

```python
print("Length of an entry is: ", len(n2020[0][0]))
print("Number of entries/observations: ", len(n2020[0]))
```

Father/Mother Education
APGAR Score (5 min)

```python
apgar5 = []         # 444-445
payment = []        # 435
father_edu = []     # 163
mother_edu = []     # 124
weight = []         # 504-507
sex = []            # 475

data = n2020[0]

#for i in data[0:100].index:
#    print(data[i])
```

```python
len(data[0:100].index)
```

```python
# reset variables
apgar5 = []         # 444-445
payment = []        # 435
father_edu = []     # 163
mother_edu = []     # 124
weight = []         # 504-507
sex = []            # 475

for i in data.index:
    apgar5.append(data[i][444-1:445+1-1])
    payment.append(data[i][435-1:435+1-1])
    father_edu.append(data[i][163-1:163+1-1])
    mother_edu.append(data[i][124-1:124+1-1])
    weight.append(data[i][504-1:507+1-1])
    sex.append(data[i][475-1:475+1-1])

df = pd.DataFrame({
    "APGAR5":apgar5, 
    "Payment":payment, 
    "Father Education": father_edu, 
    "Mother Education": mother_edu,
    "Weight": weight,
    "sex": sex})
df
```

```python
# remove illegal terms 
# initially tried to first make the original dataframe and then remove/drop bad entries. takes way too long
# this method takes MUCH shorter

print(len(apgar5))
print(len(payment))
print(len(father_edu))
print(len(mother_edu))
print(len(weight))
print(len(sex))

apgar5_clean = []
payment_clean = []
father_edu_clean = []
mother_edu_clean = []
weight_clean = []
sex_clean = []

for i in range(len(apgar5)):
    if(apgar5[i] != "99" and payment[i] != "9" and father_edu[i] != "9" and mother_edu[i] != "9" and father_edu[i] != "0" and mother_edu[i] != "0"):
        apgar5_clean.append(apgar5[i])
        payment_clean.append(payment[i])
        father_edu_clean.append(father_edu[i])
        mother_edu_clean.append(mother_edu[i])
        weight_clean.append(weight[i])
        sex_clean.append(sex[i])

print(len(apgar5_clean))
print(len(payment_clean))
print(len(father_edu_clean))
print(len(mother_edu_clean))
print(len(weight_clean))
print(len(sex_clean))

df_clean = pd.DataFrame({
    "APGAR5":apgar5_clean, 
    "Payment":payment_clean, 
    "Father Education": father_edu_clean, 
    "Mother Education": mother_edu_clean,
    "Weight": weight_clean,
    "Sex": sex_clean
    })

df_clean
```

```python
alt.data_transformers.disable_max_rows()

axis_labels = [
    '8th Grade or Less',
    '9th to 12th Grade',
    'High School or GED',
    'College Credit',
    'Associate Degree',
    'Bachelor\'s Degree',
    'Master\'s Degree',
    'Doctorate']



alt.Chart(df_clean[0:100000]).mark_point().encode(
    alt.X("Mother Education:Q", axis = alt.Axis(values = [1,2,3,4,5,6,7,8], labelFontSize = 14), scale=alt.Scale(domain=[1,8])),
    alt.Y("Father Education:Q", axis = alt.Axis(values = [1,2,3,4,5,6,7,8], labelFontSize = 14), sort=alt.Sort(order='descending'), scale=alt.Scale(domain=[1,8])),
    color = alt.Color('APGAR5:Q', scale = alt.Scale(scheme = 'spectral', domain = [0,10])),
    size = alt.Size('count()', scale=alt.Scale(range=[50,8000])),
    tooltip = ['APGAR5:Q', 'count()']
).properties(
    width = 600,
    height = 600
).configure_legend(
    # resize legend elements
    titleFontSize=18,
    labelFontSize=15
).configure_axis(
    # resize axis labels
    titleFontSize=20
).interactive()
```

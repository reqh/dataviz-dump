```python
import altair as alt
from vega_datasets import data
import pandas as pd
import numpy as np

```

```python
states = alt.topo_feature(data.us_10m.url, 'states')
statedata = pd.read_csv('us_crime_by_state.csv')
```

```python
income = pd.read_json("https://vega.github.io/vega-datasets/data/income.json")
decoder={}
for row in income.iloc():
    decoder[str.upper(row["name"])] = row["id"]
#print(decoder)
for index, row in statedata.iterrows():
    statedata.at[index, 'id'] = decoder[row['State']]

print(statedata)
```

```python
alt.Chart(states).mark_geoshape(
    stroke = 'white'
).encode(
    color = 'Crimes per 100k:Q'
).transform_lookup(
    lookup = 'id',
    from_ = alt.LookupData(statedata, 'id', ["Crimes per 100k"])
).project('albersUsa').properties(
    width = 500,
    height = 500,
    title = "Violent Crime Incidents in US per 100k Inhabitants, by State"
).configure_legend(
    titleFontSize = 14
).configure_title(
    fontSize = 20
)
```

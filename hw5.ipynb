```python
# so I'm going to add all of the nodes manually.
G2 = nx.Graph()
all_people = []
for row in data.iterrows():
    p1 = row[1]['person_id_1']
    p2 = row[1]['person_id_2']
    if p1 in all_people:
        pass
    elif p1=='G-d_1':
        G2.add_node(p1,size=30,color='red')
        all_people.append(p1)
    else:
        G2.add_node(p1)
        all_people.append(p1)
    
    if p2 in all_people:
        pass
    else:
        G2.add_node(p2)
        all_people.append(p2)

i=0
for row in data.iterrows():
    relation = row[1]['relationship_type']
    p1 = row[1]['person_id_1']
    p2 = row[1]['person_id_2']
    if i%2==0:
        dyadrelationship = relation + "/" + data.iloc[i+1,3]
        G2.add_edge(p1, p2, title=dyadrelationship)
    i+=1

net2 = Network(notebook = True, height='1000px', width='1000px')
net2.from_nx(G2)
net2.show_buttons(filter_=['nodes','edges','physics'])
net2.show("biblepeople.html")
print(len(all_people))
```
